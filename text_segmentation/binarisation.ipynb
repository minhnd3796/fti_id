```python
import cv2
import numpy as np
from PIL import Image

from matplotlib import pyplot as plt
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = fig_size[0] * 3
fig_size[1] = fig_size[1] * 3
```

```python
# img = cv2.imread('my_full_name.jpg')
img = Image.open('my_full_name.jpg')
```

```python
img = np.array(img)
resized_height = 28
img = cv2.resize(img, (round(resized_height * img.shape[1] / img.shape[0]), resized_height))
print(img.shape)
print(type(img))
plt.imshow(img)
```

```python
def convertToGrayscale(input_img, ratio=0.5):
    output_img = np.zeros((input_img.shape[0], input_img.shape[1]), dtype=np.uint8)
    for y in range(output_img.shape[0]):
        for x in range(output_img.shape[1]):
            pixel = input_img[y, x, :]
            _max = pixel[0]
            _min = pixel[0]
            if pixel[1] > pixel[0]:
                _max = pixel[1]
            else:
                _max = pixel[0]
                _min = pixel[1]
            if _max < pixel[2]:
                _max = pixel[2]
            else:
                if _min > pixel[2]:
                    _min = pixel[2]
            output_img[y, x] = int(_max * ratio + _min * (1 - ratio))
    return output_img
```

```python
grey_img = convertToGrayscale(img)
plt.imshow(grey_img, cmap='gray')
```

```python
import pbcvt
SAUVOLA = 1

binary_img = pbcvt.applyBinarisationFilter(grey_img, SAUVOLA, float(0.35), float(100))
inverted_binary_img = 255 - binary_img
plt.imshow(inverted_binary_img, cmap='gray')

print(inverted_binary_img.shape)
```

```python
mask = np.zeros_like(inverted_binary_img, dtype=np.uint8)
mask[:, 0:104] = 1
mask[:, 112:166] = 1
mask[:, 175:] = 1
plt.imshow(mask, cmap='gray')
```
